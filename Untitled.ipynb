```python
population = {
    "A": 100,
    "B": 100,
    "C": 100,
    "D": 100,
    "E": 100,
    "F": 100,
    "G": 100,
}
generations = [x for x in range(41)]
values = {label: [] for label in population.keys()}
for key in population.keys():
    values[key].append(population[key])
    
tot = sum(population.values())
for i in range(40):
    print("Generation {i}".format(i=i))
    out = subprocess.check_output('./aa -A {A} -B {B} -C {C} -D {D} -E {E} -F {F} -G {F} -rounds 50 -var 30 -var_loss -print_standing'.format(**population).split(' ')).decode('utf-8')
    l = out.split(' ')[:(tot*4//5)]
    for key in population.keys():
        population[key] = l.count(key) * 5 // 4
        values[key].append(population[key])
    print(population)
```
